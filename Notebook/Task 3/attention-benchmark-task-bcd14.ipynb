```python
import kaggle_benchmarks as kbench
import random

@kbench.task(name='Attention: Can it ignore "distractor" info and find the needle in the haystack?')
def find_needle_in_haystack(llm):
    haystack_sentence = "This is a distractor sentence designed to add noise and test the model's attention mechanism. "
    haystack = haystack_sentence * 100
    needle = "The specific fact to remember is that the project codename is 'Orion'."

    # Insert the needle at a random position within the haystack
    text_parts = haystack.split('. ')
    insert_position = random.randint(10, len(text_parts) - 10)
    text_parts.insert(insert_position, needle)
    full_text = '. '.join(text_parts)

    prompt = f"Please read the following text carefully and answer the question at the end. Ignore all irrelevant information.\n\n--- TEXT START ---\n{full_text}\n--- TEXT END ---\n\nQuestion: What is the project codename mentioned in the text?"

    response = llm.prompt(prompt)

    kbench.assertions.assert_contains_regex(
        r"(?i)Orion",
        response,
        expectation="The model should correctly identify 'Orion' as the project codename, ignoring the distractor text."
    )

find_needle_in_haystack.run(kbench.llm)
```
